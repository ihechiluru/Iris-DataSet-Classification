**Brief**

We have been asked to perform a classification analysis on the Iris dataset. In the dataset, we have been given Sepal values and Petal values.

The four features that were given are:

Sepal length


Sepal width


Petal length


Petal width



Now, Depending on these features there are basically three (3) types of flowers


They are:


Iris - Versicolor


Iris - Setosa


Iris Virginica

What are to do is to train the dataset so that with the given features, our model can predict the class our flower belongs to.

**IMPORTING LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**Importing the DataSet into our notebook**

```python
df = pd.read_csv('/content/Iris (1).csv')
```

```python
df
```

```python
from sklearn import preprocessing

from sklearn.model_selection import train_test_split

from sklearn.neighbors import KNeighborsClassifier
```

**While reading the data we want to check for the number of different Species**

```python
df['Species'].unique()
```

```python
df.isnull().values.any()
```

**Now we map int values to the label (dependent variable)**

*This would help eliminate continuity error when training the dataset*

```python
df['Species'] = df['Species'].map({'Iris-setosa' :0, 'Iris-versicolor' :1, 'Iris-virginica' :2}).astype(int) #mapping numbers
```

```python
df
```

**In the code cell below we nirmalize our data**


*This basically brings all the values of the different features on the same scale thereby optimizing our model and reducing inaccuracy*

```python
x = df.drop(['Species'], axis = 1)
y = df['Species']

MinMaxScaler = preprocessing.MinMaxScaler()

X_data_minmax = MinMaxScaler.fit_transform(x)

data = pd.DataFrame(X_data_minmax,columns=['Id','SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])

df
```

**Assigning Variables to the train_test_split**

```python
x_train,  x_test, y_train, y_test = train_test_split(data, y, test_size = 0.2, random_state = 0)
```

```python
Knn = KNeighborsClassifier(n_neighbors=7, metric= "euclidean")
```

**Training our data using the fit method**

```python
Knn.fit(x_train, y_train)
```

```python
y_pred = Knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

**Computing the confusion and accuracy scores using the KNN algorithm**

```python
confusion_matrix(y_test, y_pred)
#Confusion Matrix using KNN algorithm 
```

```python
accuracy_score(y_test, y_pred) 
#Accuracy score using KNN algorithm 
```
